# Mathplotlib

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x = (1,2,3,4,5)
y = (1,4,9,16,25)
lines = plt.plot(x,y)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.setp(lines,color = "r" , linewidth = 2.0)
plt.show()
```

```python
x = (1,2,3,4,5)
y = (1,5,7,8,9)
grafico = plt.plot(x, y, linestyle = "-.")
plt.setp(grafico, color = "b", linewidth = 2.0)
plt.show()
```

```python
import numpy as np 
def f(x):
    return x ** 2
```

```python
x1 = np.arange(8, 12)
x2 = np.arange(5, 9)
plt.subplot(2, 1, 1)
plt.title("Primer eje")
plt.plot(x1,f(x1),"ro",x2,f(x2),"b--")

plt.subplot(2, 1, 2)
plt.title("Segundo eje")
plt.plot(x1,x2,"g-.")
plt.show()
```

```python
plt.plot([4,5,6],"y-.",linewidth = 3.0)
```

```python
def funcion(x):
    return np.exp(-x)
```

```python
x1 = np.arange(1,4)
plt.plot(x1,funcion(x1),"r-.",linewidth = 2.0)
```

```python
x = np.arange(5,10)
y = np.arange(6,11)
```

```python
plt.figure(1)
plt.subplot(2, 1, 1)
plt.plot(x, y, "ro", lw = 2.0)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal",color = "red")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, y, "b--", lw = 2.0)
plt.yscale("log")
plt.xscale("log")
plt.title("Escala logartimica", color = "blue")
plt.grid(True)

plt.subplots_adjust(bottom = 0.1, hspace = 0.5)

plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
s, c = np.sin(x), np.cos(x)
plt.plot(x, s, color = "green", lw = 2.0, linestyle = "--", label = "Seno")
plt.plot(x, c, color = "red", lw = 2.0, linestyle = "-.", label = "coseno")
plt.legend(loc = "upper left")
plt.yticks(np.linspace(-1, 1, 3, endpoint = True),(["-1","0","+1"]))
plt.xlim(-4 , 4)        

ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data", 0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data", 0))

x0 = 2 * np.pi / 3
plt.plot([x0, x0], [0, np.sin(x0)], color = "blue", lw = 2.5)
plt.scatter([x0] , [np.sin(x0)], 50, color = "blue")
plt.annotate(r"$\sin ( \frac {2 \pi} {3}) = \frac {\sqrt{3}}{2}$",xy = (x0, np.sin(x0)),
            xycoords = "data", xytext = (+10, +20) , textcoords = "offset points", fontsize = 16, 
            arrowprops = dict(arrowstyle = "->"))

plt.plot([x0, x0], [0, np.cos(x0)], color = "black", lw = 2.5)
plt.scatter([x0], [np.cos(x0)], 50, color = "black")
plt.annotate(r"$\cos ( \frac {2\pi}{3} ) = -\frac {1}{2}$", xy = (x0, np.cos(x0)),
            xycoords = "data", xytext = (+10, +20),textcoords = "offset points",fontsize = 11,
            arrowprops = dict(arrowstyle = "->"))

plt.show()
```
